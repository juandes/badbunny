```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from fbprophet import Prophet
```

```python
sns.set()
```

```python
df = pd.read_csv('data/timeseries_date.csv')
```

```python
# changepoint_prior_scale=0.06
m = Prophet()
```

```python
m.fit(df)
```

```python
forecast = m.predict(df)
```
